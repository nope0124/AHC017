```python
base = []
swap = []
move = []
num = 100
with open('scores/base.txt') as f:
    lines = f.readlines()
    f.close()
    for i in range(num):
        s = "Seed_{}: ".format(str(i).zfill(4))
        lines[i] = lines[i].replace("\n", "")
        lines[i] = lines[i].replace(s, "")
        base.append(int(lines[i]))
        
with open('scores/evaluation_move.txt') as f:
    lines = f.readlines()
    f.close()
    for i in range(num):
        s = "Seed_{}: ".format(str(i).zfill(4))
        lines[i] = lines[i].replace("\n", "")
        lines[i] = lines[i].replace(s, "")
        move.append(int(lines[i]))
        
with open('scores/evaluation_swap.txt') as f:
    lines = f.readlines()
    f.close()
    for i in range(num):
        s = "Seed_{}: ".format(str(i).zfill(4))
        lines[i] = lines[i].replace("\n", "")
        lines[i] = lines[i].replace(s, "")
        swap.append(int(lines[i]))


for i in range(num):
    swap_num = swap[i] / base[i]
    move_num = move[i] / base[i]
    print("Move{}: ".format(str(i).zfill(4)) + str(move_num))
    print("Swap{}: ".format(str(i).zfill(4)) + str(swap_num) + "\n")
```

```python
N_list = []
M_list = []
D_list = []
K_list = []
P_list = []
for _ in range(num):
    with open('in/{}.txt'.format(str(_).zfill(4))) as f:
        lines = f.readlines()
        f.close()
        tmp_list = lines[0].replace("\n", "").split(" ")
        N_list.append(int(tmp_list[0]))
        M_list.append(int(tmp_list[1]))
        D_list.append(int(tmp_list[2]))
        K_list.append(int(tmp_list[3]))
        P_list.append(int(tmp_list[1]) / (int(tmp_list[2]) * int(tmp_list[3])))
```

```python
import matplotlib.pyplot as plt
move_list = []
for i in range(num):
    move_list.append(move[i] / base[i])
```

```python
plt.xlabel("D_list")
plt.ylabel("move_evaluation")
plt.scatter(D_list, move_list, s=20, cmap='Blues', vmin=0.0, vmax=1.0, linewidths=0.5, ec="b")
```

```python
plt.xlabel("M_list")
plt.ylabel("move_evaluation")
plt.scatter(move, move_list, s=20, cmap='Blues', vmin=0.0, vmax=1.0, linewidths=0.5, ec="b")
```
