```python
base = []
swap = []
move = []
num = 100
with open('scores/base.txt') as f:
    lines = f.readlines()
    f.close()
    for i in range(num):
        s = "Seed_{}: ".format(str(i).zfill(4))
        lines[i] = lines[i].replace("\n", "")
        lines[i] = lines[i].replace(s, "")
        base.append(int(lines[i]))
        
with open('scores/evaluation_move.txt') as f:
    lines = f.readlines()
    f.close()
    for i in range(num):
        s = "Seed_{}: ".format(str(i).zfill(4))
        lines[i] = lines[i].replace("\n", "")
        lines[i] = lines[i].replace(s, "")
        move.append(int(lines[i]))
        
with open('scores/evaluation_swap.txt') as f:
    lines = f.readlines()
    f.close()
    for i in range(num):
        s = "Seed_{}: ".format(str(i).zfill(4))
        lines[i] = lines[i].replace("\n", "")
        lines[i] = lines[i].replace(s, "")
        swap.append(int(lines[i]))


for i in range(num):
    swap_num = swap[i] / base[i]
    move_num = move[i] / base[i]
    print("Move{}: ".format(str(i).zfill(4)) + str(move_num))
    print("Swap{}: ".format(str(i).zfill(4)) + str(swap_num) + "\n")
```
